# 이상치(Outlier) 처리

- 데이터의 트랜드에서 벗어나는 값

```python
import pandas as pd
```

```python
df1 = pd.read_csv('03_Delivery.csv', encoding='cp949') #한글 인코더 : 'cp949'
```

```python
df1.info()
```

```python
df1.describe()
```

```python
cond1 = (df1['수량'] <0)

df2 = df1.loc[~cond1]
```

```python
df2.describe()
```

```python
cond1 = (df1['수량'] <0)

df1['수량(clean)'] = df1['수량']
df1.loc[cond1, '수량(clean)'] = 1
```

```python
df1.describe()
```

```python
cond1 = (df1['수량'] <0)
df1.loc[cond1]
```

```python
import numpy as np
```

```python
cond2 = (df1['상품번호']==999999)
df1.loc[cond2, '상품번호'] = np.nan
```

```python
cond3 = (df1['상품번호'].isnull())
df1.loc[cond3]
```
