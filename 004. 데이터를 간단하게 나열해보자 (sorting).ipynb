```python
import pandas as pd
```

```python
df1 = pd.read_csv('01_Contract_data.csv')
```

```python
df1.info()
```

```python
df1.sort_values(by='Amount_Month') #데이터의 오름차순 정렬
```

```python
df1.sort_values(by='Amount_Month', ascending=False) #데이터의 내림차순 정렬
```

```python
df1.sort_values(by=['Amount_Month','Age'], ascending=False)
#두 개 이상의 항목에 대해 내림차순 정렬
```

```python
#Amount_Month에 대해 내림차순 정렬 후, Overdue_count에 대해 오름차순 정렬
df1.sort_values(by=['Amount_Month','Overdue_count'], ascending=[False,True])
```

```python
#Amount_Month값이 높고, 연령이 높은 상위 100명의 고객 명단
df2 = df1.sort_values(by=['Amount_Month','Age'], ascending=False).head(100)
```

```python
# 4개의 항목 추출 -> Term 값에 대해 내림차순 정렬 -> 상위 50개 데이터 확인ㄴ
df1[['Sales_Type','Term','Product_Type','Overdue_count']]\
.sort_values(by='Term', ascending=False).head(50)
```
