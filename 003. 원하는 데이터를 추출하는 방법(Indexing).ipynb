```python
df2
```

```python
df1.iloc[100:200] # 원하는 범위의 데이터를 추출, iloc 함수는 특이하게 대괄호를 사용한다.
```

```python
df1.iloc[:50] # df1.head(50)과 같은 결과를 내온다.
```

```python
df1.iloc[50000:] # 50000번째부터 끝까지의 데이터 추출
```

```python
# 열 단위 데이터 추출
df1['Age'] # Series 형태로 데이터 추출
```
