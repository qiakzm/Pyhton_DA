# 기본적인 SQL 구문

- Select : 데이터를 찾을 때 / 데이터 테이블을 불러올 때
- Insert : 데이터를 집어 넣을 때
- Update : 데이터를 갱신할 때
- Delete : 데이터를 삭제할 때 (많이 안쓰인다)

```python
query1 = "SELECT * FROM sqlite_master" #데이터베이스 내 모든 table의 정보를 확인
```

```python
pd.read_sql_query(query1, connection)
```

```python
query = "SELECT * FROM 'customers'"
# *: 모든 열의 데이터를 불러옴
pd.read_sql_query(query, connection)
```

```python
query = "SELECT Company FROM 'customers'"
pd.read_sql_query(query, connection)
```

```python
query = "SELECT * FROM 'customers' WHERE Country='Canada'"
pd.read_sql_query(query, connection)
```

```python
query = "SELECT * FROM 'invoices'"
df2 = pd.read_sql_query(query, connection)
```

```python
df2.pivot_table(index='BillingCity', values='Total', aggfunc='sum')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[5,7])
sns.barplot(data=df2, y='BillingCity', x='Total', ci=None)
```

```python
df2['Date(dt)'] = pd.to_datetime(df2['InvoiceDate'])
```

```python
sns.pointplot(data=df2, x='Date(dt)', y='Total')
```

```python
sns.lineplot(data=df2, x='Date(dt)', y='Total', estimator='sum', hue='BillingCountry')
```
