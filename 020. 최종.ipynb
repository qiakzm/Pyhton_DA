```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family='AppleGothic')
```

```python
df1 = pd.read_csv('13_page.csv')
```

```python
df1_1 = df1.drop('Unnamed: 0.1', axis=1)
df1_1
```

```python
col_names = df1_1.columns.tolist()
col_names
```

```python
new_col_name = ['순위','기업명','분류','매출액_2011','영업이익_2011','영업이익률_2011',
               '매출액_2012','영업이익_2012','영업이익률','영업이익증감률']
new_col_name
```

```python
trans_name = dict(zip(col_names, new_col_name))
```

```python
df2 = df1_1.rename(columns=trans_name).iloc[4:].dropna()
```

```python
def func1(row):
    return int(row.replace(',',''))

df2['매출액_2011'].apply(func1)
```

```python
df3 = pd.melt(df2, id_vars=['순위','기업명','분류','영업이익률','영업이익증감률'])
df3
```

```python
def func2(row):
    return row.split('_')[1]

df3['연도'] = df3['variable'].apply(func2)
```

```python
def func3(row):
    return row.split('_')[0]

df3['구분'] = df3['variable'].apply(func3)
```

```python
cond1 = (df3['구분'] == '매출액')
df3_1 = df3.loc[cond1].rename(columns={'value':'매출액'})
```

```python
cond1 = (df3['구분'] == '영업이익')
df3_2 = df3.loc[cond1].rename(columns={'value':'영업이익'})
```

```python
df3_1['id'] = df3_1['기업명'] + df3_1['연도']
```

```python
df3_2['id'] = df3_2['기업명'] + df3_2['연도']
```

```python
df3_2
```

```python
df4 = pd.merge(df3_1, df3_2, on='id', suffixes=('','_drop'))
```

```python
df5 = df4[['순위','기업명','분류','연도','매출액','영업이익']]
df5
```

```python
df5['매출액(int)'] = df5['매출액'].apply(func1)
df5['영업이익(int)'] = df5['영업이익'].apply(func1)
```

```python
df5
```
