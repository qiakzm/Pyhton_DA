# 결측값 Missing value
- NaN (Not A Number) / None / NA

# 결측값 처리하는 방법
- 누락된 결측값이 있는 행을 제거
- 다른 값으로 대체 (평균이나 중앙값)

```python
import pandas as pd
```

```python
df1 = pd.read_csv('01_Contract_Data.csv')
```

```python
df1.info()
```

```python
df1.isnull() #결측값을 찾아서 True로 출력해줌
```

```python
df1.isnull().sum() #특정 항목별 결측치의 개수를 확인
```

```python
cond1 = (df1['Age'].isnull())
df1.loc[cond1]
```

```python
df2 = df1.dropna() #결측치가 있는 행을 찾아서 제거
```

```python
df2.isnull().sum()
```

```python
df1['Credit_Rank'].describe()
```

```python
# 결측치를 평균값으로 대체한 다음 새로운 항목으로 선언함.
df1['Credit_Rank(clean)'] = df1['Credit_Rank'].fillna(3.429)
```

```python
df1['Credit_Rank(clean2)'] = df1['Credit_Rank'].fillna(3.0)
```

```python
df1.head(2)
```

```python
df1.isnull().sum()
```

```python
df1.head(2)
```

```python
df1.isnull().sum()
```

```python
# 기존의 항목을 데이터프레임에서 제거

df2 = df1.drop(columns=['Credit_Rank'])
```

```python
df2.isnull().sum()
```

```python
df2['Age'].describe()
```

```python
df2['Age(clean1)'] = df2['Age'].fillna(50.024093)
```

```python
df2['Age(clean2)'] = df2['Age'].fillna(49.0)
```

```python
df3 = df2.drop(columns=['Age'])
```
