```python
import pandas as pd
```

```python
df1 = pd.read_csv('01_Contract_Data.csv')
```

```python
df1.head()
```

```python
# 고객들의 유입 채널 별 월 랜탈 비용

df1.pivot_table(index='Channel', values='Amount_Month') #채널 별 월 결제액의 평군
```

```python
#aggfunc : 특정 통계량을 계산하는 기능
df1.pivot_table(index='Channel', values='Amount_Month', aggfunc='sum')
```

```python
df1.pivot_table(index=['Gender','Channel'], values='Amount_Month', aggfunc='sum')
```

```python
df1.pivot_table(index=['Gender','Channel'], values=['Amount_Month','Overdue_count'], aggfunc=['sum','min','max'])
```

```python
df1.pivot_table(index='Sales_Type', values='Amount_Month', columns='Datetime', aggfunc='sum')
```

```python
df1['Datetime(dt)'] = pd.to_datetime(df1['Datetime'])
```

```python
df1['Year'] = df1['Datetime(dt)'].dt.year
df1['Month'] = df1['Datetime(dt)'].dt.month
```

```python
df1.pivot_table(index='Sales_Type', columns=['Year', 'Month'], values='Amount_Month', aggfunc='sum').to_excel('result1.xlsx')
```

```python
df1.pivot_table(index='Bank', values='Amount_Month', aggfunc='sum', margins=True)
```

```python
df1.pivot_table(index='Product_Type', values='Amount_Month', aggfunc='sum', margins=True, columns='Year').to_excel('result_2.xlsx')
```

```python
df3 = df1.pivot_table(index='Product_Type', values='Amount_Month', aggfunc='sum', columns='Year')
```

```python
df3['증감액'] = df3[2020]-df3[2019]
```

```python
df3['증감율'] = df3['증감액']/(df3[2020]+df3[2019])
```
