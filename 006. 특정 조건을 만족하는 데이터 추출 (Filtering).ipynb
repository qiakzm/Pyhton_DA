```python
# Contract_Type이 교체계약인 데이터만 필터링
condition1 = df1['Contract_Type']=='교체계약'
```

```python
condition1
```

```python
df1.loc[condition1]
```

```python
# Amount_Month 1500000 이상인 데이터만 필터링

cond2 = df1['Amount_Month'] >= 150000
df1.loc[cond2]
```

```python
cond2
```

```python
# Overdue_count(연체건수) 가 5회 이상인 데이터만 필터이

cond3 = (df1['Overdue_count'] >= 5)
cond3
```

```python
df2 = df1.loc[cond3]
```

```python
df2
```

```python
# Address1 가 경기도 이면서, Overdue_count 연체건수가 5회 이상인 데이터 필터링

cond4 = (df1['Address1']== '경기도')
cond5 = (df1['Overdue_count'] >= 5)
df1.loc[cond4 & cond5] #cond4 조건도 만족하면서 cond5 조건도 만족하는 데이터 추출
```

```python
#연령대가 30세 미만, 연체건수 5회 미만

cond6 = (df1['Age'] < 30)
cond7 = (df1['Overdue_count'] >= 5)
df1.loc[cond6 & cond7].head(10)
```

```python
# Address1가 경기도 이거나 충청도인 고객 데이터 필터링

cond8 = (df1['Address1']=='경기도')
cond9 = (df1['Address1']=='충청도')
df1.loc[cond8 | cond9] #cond8 조건을 만족하거나 cond9 조건을 만족하는 데이터를 모두 추출
```

```python
df1['Address1'].value_counts()
```

```python
#렌탈 유형의 고객 데이터만 추출 , 연체건수가 5회 이상인 데이터 필터링
#연체건수가 높은 상위 10명
cond10 = (df1['Sales_Type'] == '렌탈')
cond11 = (df1['Overdue_count'] >= 5)
```

```python
df1.loc[cond10 & cond11].sort_values(by='Overdue_count', ascending=False).head(10)
```

```python
# 과제 : Payment_Type 항목이 'CMS'인 데이터만 추출해 연체건수가 높은 상위 100건의 데이터 확인

cond12 = (df1['Payment_Type'] == 'CMS')
df1.loc[cond12].sort_values(by='Overdue_count', ascending=False).head(100)
```
