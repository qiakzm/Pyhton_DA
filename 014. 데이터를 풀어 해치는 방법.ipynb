```python
df3 = df1.melt(id_vars=['구분','카테고리명','자재그룹','자재그룹명','제품명','단위','판매','안전재고',' 분류'])\
.rename(columns={'variable':'일자','value':'매출액'})
```

```python
df3['일자(dt)'] = pd.to_datetime(df3['일자'])
df3['연도'] = df3['일자(dt)'].dt.year
df3['월'] = df3['일자(dt)'].dt.month
```

```python
df3
```

```python
df3.pivot_table(index=['연도','월'], values='매출액', aggfunc='sum', columns='제품명')
```

```python
sns.barplot(data=df3, x='월', y='매출액', ci=None)
```

```python
sns.barplot(data=df3, x='월', y='매출액', hue='연도', ci=None)
```

```python
cond1 = (df3['월'] == 12)
df4 = df3.loc[cond1].pivot_table(index='제품명', values='매출액', columns='연도', aggfunc='sum')
```

```python
df4['증감액'] = df4[2020]-df4[2019]
```

```python
df4['증감율'] = df4['증감액'] / (df4[2020]+df4[2019])
```

```python
df4.sort_values(by='증감율', ascending=False)
```
