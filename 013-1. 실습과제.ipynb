```python
import pandas as pd
```

```python
df1 = pd.read_csv('01_Contract_Data.csv')
```

```python
df1['Datetime(dt)'] = pd.to_datetime(df1['Datetime'])
```

```python
df1['주문요일'] = df1['Datetime(dt)'].dt.dayofweek
```

```python
df1['주문요일']
```

```python
df1.pivot_table(index=['주문요일','Product_Type'], values='Amount_Month', aggfunc='sum')
```

```python
df3 = df1.pivot_table(index=['주문요일','Product_Type'], values='Amount_Month', aggfunc='sum').sort_values(by='주문요일')
```

```python
df3.rename({0 : 'Monday',
            1 : 'Tuesday',
            2 : 'Wednesday',
            3 : 'Thursday',
            4 : 'Friday',
            5 : 'Saturday',
            6 : 'Sunday'}, axis='index')
```
